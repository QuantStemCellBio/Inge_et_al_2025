```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
from re import match
import os
import numpy as np
import seaborn as sns
import math

ingeo_colours = ["#0077BB","#33BBEE","#009988","#EE7733","#CC3311","#EE3377","#BBBBBB", "#A020F0"]
```

```python
df = pd.read_csv("full_pos_sub.csv")
```

```python
df['SOX17+'] = df['SOX17+FOXC1-'] + df['SOX17+FOXC1+']
```

```python
df = df[["Condition", "SOX17+"]]


order= list(["A100B10_NTC_50nM",
             "A100_NTC_50nM",
             "A100_MESP1_50nM",
             "A100_MESP2_50nM",
             "A100_FOXC1_50nM",
             "A100_NKX2-5_50nM",
             "A100_GATA3_50nM",
             "A100_HAND1_50nM"])
```

```python

plt.figure(figsize=(3,2.5), dpi=100)
sns.set_style("ticks")

y_axis_limit=100

ax = sns.barplot(data=df, x='Condition', y='SOX17+', 
                capsize=0.3, errwidth=1.5, dodge=False, palette=[ingeo_colours[0]], order=order)
sns.despine(ax=ax)
plt.ylim(0,100)
plt.xticks(rotation=90)
```
