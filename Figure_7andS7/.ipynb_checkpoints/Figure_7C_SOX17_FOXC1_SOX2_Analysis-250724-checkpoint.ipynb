```python
ingeo_colours = ["#0077BB","#33BBEE","#009988","#EE7733","#CC3311","#EE3377","#BBBBBB", "#A020F0"]
```

```python
df_1 = pd.read_csv('OI_202407010_BG_CASA_SOX17647_FOXC1568_SOX2488_001/feature_data.csv')
df_2 = pd.read_csv('OI_202407014_BG_CASA_SOX17647_FOXC1568_SOX2488_001/feature_data.csv')
```

```python
def split_condition(df):
    # Split 'Condition' column into four parts
    split_columns = df['Condition'].str.split('_', n=3, expand=True)
    # Assign the split parts to new columns
    df['Treatment'], df['siRNA'], df['Concentration'], df['Repeat'] = split_columns[0], split_columns[1], split_columns[2], split_columns[3]
    return df
```

```python
df_1 = split_condition(df_1)
df_2 = split_condition(df_2)
```

```python
df_1 = df_1.loc[df_1['area']>200]
df_2 = df_2.loc[df_2['area']>200]
```

```python
def sample_equally(dataframe, class_, sample_n):
    sample_list = list()
    for i in dataframe[class_].unique():
        df_ = dataframe.loc[dataframe[class_]==i]
        df_ = df_.sample(sample_n)
        sample_list.append(df_)
    df_ = pd.concat(sample_list)
    return df_
        
```

```python
df_1_threshold = sample_equally(df_1, 'Condition', 1000)
df_2_threshold = sample_equally(df_2, 'Condition', 1000)

df_1_threshold = df_1_threshold.loc[df_1_threshold['siRNA'] =='NTC']
df_2_threshold = df_2_threshold.loc[df_2_threshold['siRNA'] =='NTC']

fig, axes = plt.subplots(2, 2, figsize=(15, 15))

axes = axes.flatten()

sns.kdeplot( ax=axes[0],
    data=df_1_threshold, 
    x="mean_intensity_647",  hue="Treatment"
)
axes[0].axvline(x=2100, linewidth=0.3, color='r')

sns.kdeplot( ax=axes[1],
    data=df_1_threshold, 
    x="mean_intensity_mCherry",  hue="Treatment"
)
axes[1].axvline(x=800, linewidth=0.3, color='r')


sns.kdeplot( ax=axes[2],
    data=df_2_threshold, 
    x="mean_intensity_647",  hue="Treatment"
)
axes[2].axvline(x=2100, linewidth=0.3, color='r')


sns.kdeplot( ax=axes[3],
    data=df_2_threshold, 
    x="mean_intensity_mCherry",  hue="Treatment"
)
axes[3].axvline(x=1000, linewidth=0.3, color='r')


plt.tight_layout()
plt.show()
```

```python
T_mch = 1000
T_647 = 4000

def calc_data_pos(filtered_Sel, T_mch, T_647):

    # Combined cell type status
    subset = filtered_Sel
    groups = "Condition"



    subset.loc[(subset['mean_intensity_647'] >= T_647) & (subset['mean_intensity_mCherry'] >= T_mch), 'HAND1_GATA6_status'] = 'SOX17+FOXC1+'  
    subset.loc[(subset['mean_intensity_647'] <= T_647) & (subset['mean_intensity_mCherry'] >= T_mch), 'HAND1_GATA6_status'] = 'SOX17-FOXC1+'  
    subset.loc[(subset['mean_intensity_647'] >= T_647) & (subset['mean_intensity_mCherry'] <= T_mch), 'HAND1_GATA6_status'] = 'SOX17+FOXC1-'  
    subset.loc[(subset['mean_intensity_647'] <= T_647) & (subset['mean_intensity_mCherry'] <= T_mch), 'HAND1_GATA6_status'] = 'SOX17-FOXC1-' 

    celltypes = subset.groupby(["Condition", 'HAND1_GATA6_status']).count()
    celltypes = pd.DataFrame(celltypes['label']).reset_index().pivot(index=groups, columns='HAND1_GATA6_status', values='label')
    H1G6_status = celltypes.fillna(0)

    counts = filtered_Sel.groupby(["Condition"]).count()

    data = [counts["label"],
            H1G6_status["SOX17+FOXC1-"],
            H1G6_status["SOX17-FOXC1+"],
            H1G6_status["SOX17+FOXC1+"],
            H1G6_status["SOX17-FOXC1-"]

           ]

    headers = ["Count", 
               "SOX17+FOXC1-",
               "SOX17-FOXC1+",
               "SOX17+FOXC1+",
               "SOX17-FOXC1-"
              ]

    data_pos = pd.concat(data, axis=1, keys=headers)
    data_pos = data_pos.reset_index()
    data_pos = data_pos.rename(columns={'index': 'Condition'})

    data_pos[['Treatment', 'siRNA','Concentration','Repeat']] = data_pos['Condition'].str.split('_', 3, expand=True)


    data_pos["SOX17+FOXC1-"] = data_pos["SOX17+FOXC1-"]/data_pos["Count"]*100
    data_pos["SOX17-FOXC1+"] = data_pos["SOX17-FOXC1+"]/data_pos["Count"]*100
    data_pos["SOX17+FOXC1+"] = data_pos["SOX17+FOXC1+"]/data_pos["Count"]*100
    data_pos["SOX17-FOXC1-"] = data_pos["SOX17-FOXC1-"]/data_pos["Count"]*100

    return data_pos
```

```python
data_pos_1 = calc_data_pos(df_1, 1000, 1300)
data_pos_2 = calc_data_pos(df_2, 1000, 2100)
```

```python
data_pos_1['Condition'] = data_pos_1[['Treatment', 'siRNA','Concentration']].agg('_'.join, axis=1)
data_pos_2['Condition'] = data_pos_2[['Treatment', 'siRNA','Concentration']].agg('_'.join, axis=1)
```

```python
data_pos_1 = data_pos_1.loc[data_pos_1['Concentration']=='50nM']
```

```python
data_pos_2.loc[(data_pos_2['Treatment']=='A100B10')&(data_pos_2['siRNA']=='NTC')&(data_pos_2['Repeat']=='R1'),'Repeat'] = 'R3'
data_pos_2.loc[(data_pos_2['Treatment']=='A100B10')&(data_pos_2['siRNA']=='NTC')&(data_pos_2['Repeat']=='R2'),'Repeat'] = 'R4'

data_pos_2.loc[(data_pos_2['Treatment']=='A100')&(data_pos_2['siRNA']=='NTC')&(data_pos_2['Repeat']=='R1'),'Repeat'] = 'R3'
data_pos_2.loc[(data_pos_2['Treatment']=='A100')&(data_pos_2['siRNA']=='NTC')&(data_pos_2['Repeat']=='R2'),'Repeat'] = 'R4'

full_pos = pd.concat([data_pos_1, data_pos_2])
full_pos.to_csv('full_pos.csv')

full_pos_ctr = full_pos.loc[(full_pos['Treatment']=='A100B10')&(full_pos['siRNA']=='NTC')]
full_pos_test = full_pos.loc[(full_pos['Treatment']=='A100')]
full_pos_sub = pd.concat([full_pos_test, full_pos_ctr])

full_pos_sub['SOX17+'] = full_pos_sub['SOX17+FOXC1-'] + full_pos_sub['SOX17+FOXC1+']
full_pos_ctr['SOX17+'] = full_pos_ctr['SOX17+FOXC1-'] + full_pos_ctr['SOX17+FOXC1+']
```

```python
def create_stack_bar_plot(
    df,
    df_error_bar=None,
    x_figSize=2.5,
    y_figSize=2.5,
    y_label=None,
    y_axis_start=0,
    y_axis_limit=None,
    color_pal=sns.color_palette(palette="Blues_r"),
    bar_width=0.8,
    x_label =None
):

    fig, ax = plt.subplots(figsize=(x_figSize, y_figSize))

    sns.set(style="ticks")

    ax = df.plot(
        kind="bar",
        stacked=True,
        color=color_pal,
        width=bar_width,
        ax=ax,
        yerr=df_error_bar,
        capsize=4,
    )
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)

    sns.despine(ax=ax)
    ax.xaxis.set_tick_params(width=1)
    ax.yaxis.set_tick_params(width=1)
    ax.tick_params(axis="both", which="major", pad=1)
    ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.setp(ax.spines.values(), linewidth=1)

    if not y_axis_limit == None:
        ax.set_ylim(top=y_axis_limit)

    handles, labels = ax.get_legend_handles_labels()

    ax.legend(
        reversed(handles), reversed(labels), bbox_to_anchor=(1, 1), loc="upper left"
    )
```
