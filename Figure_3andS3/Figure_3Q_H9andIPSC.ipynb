```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ingeo_colours = ["#0077BB","#33BBEE","#009988","#EE7733","#CC3311","#EE3377","#BBBBBB","#5D54A4"]
```

```python
df = pd.read_csv('ParameterData_Main.txt', delimiter = "\t")
```

```python
Cells = df.loc[df['Cell']==1]
```

```python
Cells.columns
```

```python
subset = Cells.loc[Cells['Well '].isin([15, 27, 39, 20, 32, 44, 51, 63, 75])]
```

```python
subset.loc[(subset['Well ']==15), 'Condition'] = 'H1_R1'
subset.loc[(subset['Well ']==27), 'Condition'] = 'H1_R2'
subset.loc[(subset['Well ']==39), 'Condition'] = 'H1_R3'

subset.loc[(subset['Well ']==20), 'Condition'] = 'H9_R1'
subset.loc[(subset['Well ']==32), 'Condition'] = 'H9_R2'
subset.loc[(subset['Well ']==44), 'Condition'] = 'H9_R3'

subset.loc[(subset['Well ']==51), 'Condition'] = 'CL4_R1'
subset.loc[(subset['Well ']==63), 'Condition'] = 'CL4_R2'
subset.loc[(subset['Well ']==75), 'Condition'] = 'CL4_R3'
```

```python
sample = subset.sample(n=10000)
```

```python
sample = subset.sample(n=10000)
sample[['CellLine', 'Rep']] = sample['Condition'].str.split('_', n=1, expand=True)
```

```python
#sample.to_csv("H1.csv")
```

```python

FOXC1T = 1200
SOX17T = 600


sample.loc[((subset["Mean Intensity GFP"]>FOXC1T) &(subset["Mean Intensity Alexa 647"]>SOX17T)), 'Population'] = 'FOXC1+SOX17+'
sample.loc[((subset["Mean Intensity GFP"]<FOXC1T) &(subset["Mean Intensity Alexa 647"]>SOX17T)), 'Population'] = 'FOXC1-SOX17+'
sample.loc[((subset["Mean Intensity GFP"]>FOXC1T) &(subset["Mean Intensity Alexa 647"]<SOX17T)), 'Population'] = 'FOXC1+SOX17-'
sample.loc[((subset["Mean Intensity GFP"]<FOXC1T) &(subset["Mean Intensity Alexa 647"]<SOX17T)), 'Population'] = 'FOXC1-SOX17-'

hue_order = ["FOXC1+SOX17+", "FOXC1-SOX17+", "FOXC1-SOX17-", "FOXC1+SOX17-"]

Cell_Line = 'H1'
prop = sample.loc[sample['CellLine']==Cell_Line].groupby('Population').count()
prop = prop.reset_index()
prop['Percetage'] = (prop['Object ID']/sum(prop['Object ID']))*100


plt.figure(figsize=(2, 2), dpi=100)

sns.scatterplot(data=sample.loc[sample['CellLine']==Cell_Line], x="Mean Intensity GFP", y="Mean Intensity Alexa 647", hue="Population", 
                alpha=0.2, palette=[ingeo_colours[7],
                                     ingeo_colours[0],
                                     ingeo_colours[6],
                                     ingeo_colours[4]],   # or any Seaborn predefined palette
                hue_order=hue_order).set(yscale="log", xscale="log", xlim=(0,10000), ylim=(0,3000))

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

x1 = round(prop['Percetage'][0], 0)
x2 = round(prop['Percetage'][1], 0)
x3 = round(prop['Percetage'][2], 0)
x4 = round(prop['Percetage'][3], 0)

plt.text(700, 2500, f'{x1}%', fontsize=12, ha='center', va='center', color='black')
plt.text(7200, 2500, f'{x2}%', fontsize=12, ha='center', va='center', color='black')
plt.text(700, 300, f'{x3}%', fontsize=12, ha='center', va='center', color='black')
plt.text(7200, 300, f'{x4}%', fontsize=12, ha='center', va='center', color='black')


plt.savefig("H1_proportions.pdf", dpi=300, bbox_inches='tight')
```
